```python
# Get and load data
import pandas as pd
def get_data():
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    html = pd.read_html(url, header = 0)
    df = html[0]
    return df
    
df = get_data()
df.head()
```

```python
# Examine the sectors
sector_unique = df['GICS Sector'].unique()
len(sector_unique)
```

```python
# Group by sector
sector = df.groupby('GICS Sector')
sector.first()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create a flexible plot
def price_plot(symbol):
    df = pd.DataFrame(data[symbol].Close)
    df['Date'] = df.index
    plt.fill_between(df.Date, df.Close, color="green", alpha=0.5)
    plt.plot(df.Date, df.Close, color="green", alpha=0.9)
    plt.title(symbol, fontweight='bold')
    plt.xticks(rotation=90)
    plt.xlabel("Date")
    plt.ylabel("Closing price")
    return plt.show()
```

```python
print(price_plot('ABT'))
```

```python
for i in list(df.Symbol)[:5]:
    price_plot(i)
```
